```python
import matplotlib.pyplot as plt 

#Lager tom liste for x-verdier.
x=[]

# Legger til x-verdier.
for n in range(-5,5):
    x.append(n)
```

```python
#Lager tom liste for y-verdier.
y=[]

# Legger til y-verdier.
for n in range(-5,5):
    y.append(n**2)
```

```python
print("x-verdier:", x)
print("y-verdier:", y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('tid (s)')                                       #Navn på x-akse.
plt.ylabel('fart (m/s)')                                    #Navn på y-akse.
plt.title('Verdens kuleste funksjon, f(x)=x^2')             #Navn på figuren.
plt.grid(True)                                              #Ta med rutenett.
plt.show()
```

```python
plt.plot(x,y, '--')                                         # '-' for stiplet linje.
plt.xlabel('tid (s)')                                       #Navn på x-akse.
plt.ylabel('fart (m/s)')                                    #Navn på y-akse.
plt.title('Verdens kuleste funksjon, f(x)=x^2')             #Navn på figuren.
plt.grid(True)                                              #Ta med rutenett.
plt.show()
```

```python
plt.plot(x,y, 'g--')                                         # '-' for stiplet linje. g for grønn.
plt.xlabel('tid (s)')                                       #Navn på x-akse.
plt.ylabel('fart (m/s)')                                    #Navn på y-akse.
plt.title('Verdens kuleste funksjon, f(x)=x^2')             #Navn på figuren.
plt.grid(True)                                              #Ta med rutenett.
plt.show()
```
